```python
from pdfminer.high_level import extract_text

#         #################################################################################


def order_scraper(file1, file2):
    fileText1 = extract_text(file1)
    fileText2 = extract_text(file2)
    list1 = []
    list2 = []
    
#     slices catalog numbers out of string
    fileText1 = fileText1.split("Item",1)[1]
    fileText1 = fileText1.split("Akoya",1)[0]
    
    fileText2 = fileText2.split("Item",1)[1]
    fileText2 = fileText2.split("Akoya",1)[0]

#     Goes through product codes line by line. Appends them to list. Prints codes line by line. Ignores line if it is blank.
    for line in fileText1.splitlines():
        if line == "":
            continue
        
        else:
            list1.append(line)
            print(line)
    print("------------")
        
            
    for line in fileText2.splitlines():
        if line == "":
            continue
        
        else:
            list2.append(line)
            print(line)
    print("------------")
    
#     If item lists do not match, parse through shipping list to find absent items
    if list2 != list1:
        for i in range(0, len(list2)-1):
            if list2[i] in list1:
                print("Item " + str(list2[i]) + " has found a match")
                continue
            else:
                print("Missing item: " + list2[i])
    else:
        print("All items present")
            
    
#         #################################################################################
        
    
order_scraper("/Users/wesleyalejandro/Documents/sample-a.pdf", "/Users/wesleyalejandro/Documents/sample-b.pdf")
       
            
        

```
